Задачи, которые сформулировал руководитель:

2) провести анализ охвата аудитории (reach/reach_subscribers);

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dm-fedorov/python_basic/master/data/vkontakte_group_01_2016-08-01_2020-03-15.csv")
df['Дата'] = pd.to_datetime(df['Дата'])
df.head(5)
```

```python
df2020 = df[(df['Критерий'] == 'views') & (pd.Timestamp(2020, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2021, 1, 1))]
df2019 = df[(df['Критерий'] == 'views') & (pd.Timestamp(2019, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2020, 1, 1))]
df2018 = df[(df['Критерий'] == 'views') & (pd.Timestamp(2018, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2019, 1, 1))]
df2017 = df[(df['Критерий'] == 'views') & (pd.Timestamp(2017, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2018, 1, 1))]
views = [df2017['Значение'].sum(), df2018['Значение'].sum(), df2019['Значение'].sum(), df2020['Значение'].sum()]
df2020 = df[(df['Критерий'] == 'visitors') & (pd.Timestamp(2020, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2021, 1, 1))]
df2019 = df[(df['Критерий'] == 'visitors') & (pd.Timestamp(2019, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2020, 1, 1))]
df2018 = df[(df['Критерий'] == 'visitors') & (pd.Timestamp(2018, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2019, 1, 1))]
df2017 = df[(df['Критерий'] == 'visitors') & (pd.Timestamp(2017, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2018, 1, 1))]
visitors = [df2017['Значение'].sum(), df2018['Значение'].sum(), df2019['Значение'].sum(), df2020['Значение'].sum()]
df2 = pd.DataFrame({'Года':[2017,2018,2019,2020],'Просмотры в год':views,'Посещения в год': visitors})
a = df2.plot.bar(x='Года')
print('Статистика по посещениям и просмотрам:',a)
```

```python
df2020 = df[(df['Критерий'] == 'reach') & (pd.Timestamp(2020, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2021, 1, 1))]
df2019 = df[(df['Критерий'] == 'reach') & (pd.Timestamp(2019, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2020, 1, 1))]
df2018 = df[(df['Критерий'] == 'reach') & (pd.Timestamp(2018, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2019, 1, 1))]
df2017 = df[(df['Критерий'] == 'reach') & (pd.Timestamp(2017, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2018, 1, 1))]
reach = [df2017['Значение'].sum(), df2018['Значение'].sum(), df2019['Значение'].sum(), df2020['Значение'].sum()]
df2020 = df[(df['Критерий'] == 'reach_subscribers') & (pd.Timestamp(2020, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2021, 1, 1))]
df2019 = df[(df['Критерий'] == 'reach_subscribers') & (pd.Timestamp(2019, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2020, 1, 1))]
df2018 = df[(df['Критерий'] == 'reach_subscribers') & (pd.Timestamp(2018, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2019, 1, 1))]
df2017 = df[(df['Критерий'] == 'reach_subscribers') & (pd.Timestamp(2017, 1, 1) <= df['Дата']) &
            (df['Дата'] < pd.Timestamp(2018, 1, 1))]
reach_subscribers = [df2017['Значение'].sum(), df2018['Значение'].sum(), df2019['Значение'].sum(), df2020['Значение'].sum()]
df3 = pd.DataFrame({'Года':[2017,2018,2019,2020],'Полный охват':reach,'Охват подписчиков': reach_subscribers})
b = df3.plot.bar(x='Года')
print('Статистика по охвату аудитории:',b)
```
